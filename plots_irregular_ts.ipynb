### Aligned Multivariate Time Series

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure for plotting
plt.figure(figsize=(10, 6))

# Simulate time points
time = np.linspace(0, 10, 21)

# Generate aligned periodic signals (sine wave)
sensor_1 = np.sin(time)
sensor_2 = np.sin(time + np.pi / 4)  # Phase shift by pi/4 for distinction
sensor_3 = np.sin(time + np.pi / 2)  # Phase shift by pi/2 for further distinction

# Plot each sensor with distinct colors and labels
plt.plot(time, sensor_1, 'b-', label='Sensor 1', marker='o', linewidth=2)
plt.plot(time, sensor_2, 'g-', label='Sensor 2', marker='o', linewidth=2)
plt.plot(time, sensor_3, 'r-', label='Sensor 3', marker='o', linewidth=2)

# Customize the plot
plt.title('Aligned Time Series - No Missing')
plt.ylabel('Value')
plt.xlabel('Time')

# Set grid to align with the time points and y values
plt.grid(True)
plt.xticks(time)  # Ensure x-ticks are placed exactly at the data points
plt.yticks(np.arange(-1, 1.1, 0.5))  # Adjust y-ticks for better alignment with sine wave range

# Add legend
plt.legend(loc='upper right')

# Show plot
plt.show()
```
